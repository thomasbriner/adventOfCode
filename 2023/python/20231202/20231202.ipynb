```python
import pandas as pd
```

```python
#input = 'test_input1.txt'
input = 'input.txt'
```

```python
colors = ['red', 'green', 'blue']


with open(input) as f:
    games = []
    lines = f.readlines()
    for line in lines:
        parts = line.split(":")
        gameNr = int(parts[0].replace("Game","").strip())
        rounds = parts[1].split(";")
        for round in rounds:
            dices = {}
            round = round.replace("\n","")
            dicesWithColors = round.split(",")
            for dicesOfSingleColor in dicesWithColors:
                for color in colors:
                    if color in dicesOfSingleColor:
                        dices[color] = int(dicesOfSingleColor.replace(color,"").strip())
            games.append((gameNr, dices))    

games        
```

```python
available_dices = {}
available_dices[colors[0]] = 12
available_dices[colors[1]] = 13
available_dices[colors[2]] = 14
```

```python
def is_valid_round(dices):
    valid = True
    for color in colors:
        if color in dices and dices[color] > available_dices[color]:
            valid = False
    return valid
```

```python
illegal_games = [round for round in games if not is_valid_round(round[1])]
illegal_games
```

```python
def set_of_game_numbers(game_records):
    return set(map(lambda r: r[0], game_records))
```

```python
all_games = set_of_game_numbers(games)
all_games
```

```python

illegal_game_set =set_of_game_numbers(illegal_games)
illegal_game_set
```

```python
legal_games = all_games.difference(illegal_game_set)
legal_games
```
