```python
def get_max_nof_cols(lines):
    max_cols = 0
    for line in lines:
        if len(line) > max_cols:
            max_cols = len(line)
    return max_cols
```

```python
class GearStar:
    def __init__(self, first_pos):
        self.pos = first_pos

    def __str__(self):
        return 'GGGG'
        return f"GearStar: pos:{self.pos}"

    def get_adjacent_positions(self):
        adj = []
        adj.append((self.pos[0]-1, self.pos[1]-1))
        adj.append((self.pos[0]-1, self.pos[1]-0))
        adj.append((self.pos[0]-1, self.pos[1]+1) )       
        adj.append((self.pos[0], self.pos[1]-1))
        adj.append((self.pos[0], self.pos[1]+1))
        adj.append((self.pos[0]+1, self.pos[1]-1))
        adj.append((self.pos[0]+1, self.pos[1]-0))
        adj.append((self.pos[0]+1, self.pos[1]+1))
        return adj

    def __repr__(self):
        return self.__str__()

```

```python
class ElfNumber:
  def __init__(self, first_pos, value):
    self.first_pos_row = first_pos[0]
    self.first_pos_col = first_pos[1]
    self.value = value

  def __str__(self):
    return 'E'+"{:<3}".format(str(self.value))
    return f"ElfNumber: pos: ({self.first_pos_row},{self.first_pos_col}): {self.value}"
  
  def addValue(self, new_value):
    self.value = self.value * 10 + new_value
  
  def get_adjacent_positions(self):
    nof_places = math.floor(math.log10(self.value))
    adj = []
    # row above
    row_above = self.first_pos_row-1
    adj.append((row_above, self.first_pos_col-1))
    for i in range(0,nof_places+2):
      # add one too much
      adj.append((row_above, self.first_pos_col+i))
    # row itself
    adj.append((self.first_pos_row, self.first_pos_col-1))
    adj.append((self.first_pos_row, self.first_pos_col+nof_places+1))
    # row below
    row_below = self.first_pos_row+1
    adj.append((row_below, self.first_pos_col-1))
    for i in range(0,nof_places+2):
      # add one too much
      adj.append((row_below, self.first_pos_col+i))
    return adj
  
  def __repr__(self):
    return self.__str__()

```

```python
def parse_line(row_num, row, playground,numbers, stars):
    act_num = None
    for col, c in enumerate(row):
        if c.isdigit():
            if act_num == None:
                act_num = ElfNumber((row_num, col+1), int(c))
                playground[row_num][col+1] = act_num
            else:
                act_num.addValue(int(c))
                playground[row_num][col+1] = act_num
        else:
            if act_num != None:
                numbers.append(act_num)
                act_num = None
            if c == '*':
                star = GearStar((row_num, col+1))
                playground[row_num][col+1] = star
                stars.append(star)
    if act_num != None:
        numbers.append(act_num)
        playground[row_num][len(row)] = act_num

```

```python
def print_playground(playground):
    print()
    print()
    for row in playground:
        print(row)
    print()
    print()
```

```python
with open(input) as f:
    lines = f.readlines()
    lines =  [line.replace("\n", "") for line in lines]
    nof_cols = get_max_nof_cols(lines) + 2
    nof_rows = len(lines) +2

    numbers = []
    stars = []
    playground = []
    for i in range(1,nof_rows):
        playground.append([None]*nof_cols)

    for i in range(0,nof_rows-2):
        parse_line(i+1,lines[i], playground, numbers, stars)


    print_playground(playground)
    print(numbers)
    print(stars)

    sum = 0
    for star in stars:
        positions = star.get_adjacent_positions()
        print(positions)
        adj_numbers = []
        for position in positions:
            item = playground[position[0]][ position[1]]
            if isinstance(item, ElfNumber):
                adj_numbers.append(item)
        adj_numbers = list(set(adj_numbers))
        if (len(adj_numbers) == 2):
            sum += adj_numbers[0].value * adj_numbers[1].value
        print(adj_numbers)
    print(sum)




```
